# Cleaning Rate Column


```python
df['rate'].unique()
```

```python
#Removing "NEW" , "-" and "/5" from Rate Column
def handterate(value):
    if (value=='NEW'or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)

df['rate']=df['rate'].apply(handterate)
df['rate'].head
```

```python
df.head()
```

```python
df.drop(['city'],axis=1)
```

```python
#filling null value with mean
df['rate'].fillna(df['rate'].mean,inplace=True)
df['rate'].isnull().sum()
```

```python
df.head()
```

```python
#Dropping Null Values
df.dropna(inplace = True)
df.head()
```

```python
df['cost']
```

```python
def handlecomma(value):
    value=str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
df['cost']=df['cost'].apply(handlecomma)
df['cost'].unique
```

```python
rest_type=df['rest_type'].value_counts(ascending=False)
rest_type
```

```python
rest_type_lessthan1000=rest_type[rest_type<1000]
rest_type_lessthan1000
```

```python
def rest(value):
    if value in rest_type_lessthan1000:
        return 'others'
    else:
        return value
df['rest_type']=df['rest_type'].apply(rest)
df['rest_type']
```

```python
cuisines=df['cuisines'].value_counts(ascending=False)
cuisines
```

```python
cuisines_lessthan100=cuisines[cuisines<100]
cuisines_lessthan100.count()
```

```python
def cuisin(value):
    if value in cuisines_lessthan100:
        return 'others'
    else:
        return value
df['cuisines']=df['cuisines'].apply(cuisin)
df['cuisines'].head()
```

```python
location=df['location'].value_counts(ascending=True)
location
```

```python
locationlessthan500=location[location<500]
locationlessthan500.count()
```

```python
def location(value):
    if value in locationlessthan500:
        return 'others'
    else:
        return value
df['location']=df['location'].apply(location)
df['location']
```
